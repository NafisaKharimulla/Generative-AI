## 🧹 Text Cleaning

We remove:
- Punctuation
- Numbers
- Special characters
- Convert text to lowercase

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

cleaned_reviews = [clean_text(r) for r in reviews]
cleaned_reviews
```

## 🚫 Stop Words

Stop words are common words (is, the, and) that add little meaning.

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_tokens = [
    [word for word in review if word not in stop_words]
    for review in tokens
]

filtered_tokens
```

## 🌱 Stemming vs Lemmatization

- **Stemming**: Cuts words to root form (playing → play)
- **Lemmatization**: Converts words to dictionary form (better accuracy)

We will demonstrate both.

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed = [[stemmer.stem(w) for w in review] for review in filtered_tokens]
lemmatized = [[lemmatizer.lemmatize(w) for w in review] for review in filtered_tokens]

stemmed, lemmatized
```

## 📚 Corpus

A **corpus** is a large collection of processed text documents.

```python
corpus = [" ".join(review) for review in lemmatized]
corpus
```

```python

# FULL NLP PREPROCESSING PIPELINE


import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk import pos_tag, ne_chunk

# Download required resources (run once)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')

#  Input: Google Review

review = "I recently visited this restaurant in Bangalore. The food was amazing, but the service was very slow and disappointing."

print("Original Review:\n", review)


#  1. Lowercasing

review_lower = review.lower()


#  2. Remove Punctuation

review_no_punct = review_lower.translate(str.maketrans('', '', string.punctuation))

#  3. Tokenization

tokens = word_tokenize(review_no_punct)


#  4. Remove Stopwords

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]


#  5. Stemming

stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]


#  6. POS Tagging

pos_tags = pos_tag(tokens)


#  7. Named Entity Recognition (NER)

ner_tree = ne_chunk(pos_tags)


#  Outputs

print("\nLowercase:\n", review_lower)
print("\nNo Punctuation:\n", review_no_punct)
print("\nTokens:\n", tokens)
print("\nAfter Stopword Removal:\n", filtered_tokens)
print("\nStemmed Tokens:\n", stemmed_tokens)
print("\nPOS Tags:\n", pos_tags)
print("\nNamed Entities:\n", ner_tree)
```

```python

# WORD2VEC TRAINING (LARGE CORPUS)


import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
import string

#  Download dataset
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('punkt_tab')


#  Load Large Corpus

sentences = []

for fileid in gutenberg.fileids():
    text = gutenberg.raw(fileid)

    # Preprocessing
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Tokenize into sentences
    for sent in text.split('\n'):
        words = word_tokenize(sent)
        if len(words) > 2:
            sentences.append(words)

print("Total sentences:", len(sentences))


#  Train Word2Vec Model

model = Word2Vec(
    sentences,
    vector_size=100,   # embedding size
    window=5,          # context window
    min_count=2,       # ignore rare words
    workers=4
)


# Test Model

print("\nVector for 'king':\n", model.wv['king'][:10])

print("\nMost similar words to 'king':")
print(model.wv.most_similar('king'))
```

```python

# COSINE SIMILARITY COMPARISON


import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
import string

# Download data
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('punkt_tab')


#  Prepare corpus

sentences = []

for fileid in gutenberg.fileids():
    text = gutenberg.raw(fileid)
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))

    for sent in text.split('\n'):
        words = word_tokenize(sent)
        if len(words) > 2:
            sentences.append(words)


#  Train Word2Vec

model = Word2Vec(sentences, vector_size=100, window=5, min_count=2, workers=4)


#  Compare similarities

pairs = [
    ("good", "excellent"),   # positive-positive
    ("bad", "terrible"),     # negative-negative
    ("good", "bad"),         # positive-negative
    ("happy", "sad")         # opposite sentiment
]

print("Cosine Similarity Results:\n")

for w1, w2 in pairs:
    try:
        similarity = model.wv.similarity(w1, w2)
        print(f"{w1} ↔ {w2} : {similarity:.3f}")
    except KeyError:
        print(f"{w1} or {w2} not in vocabulary")
```

```python

# NER ON MULTIPLE CUSTOMER REVIEWS


import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk

#  Download required resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')


#  Customer Feedback Sentences

reviews = [
    "I ordered a phone from Amazon and it was delivered in Bangalore.",
    "The service at Starbucks in Mumbai was excellent.",
    "Flipkart delivered my laptop late in Delhi.",
    "I had a bad experience with Uber in Hyderabad.",
    "The food at Domino's Pizza in Chennai was amazing."
]


#  Perform NER

for review in reviews:
    print("\nReview:", review)

    # Tokenize
    tokens = word_tokenize(review)

    # POS Tagging
    pos_tags = pos_tag(tokens)

    # Named Entity Recognition
    ner_tree = ne_chunk(pos_tags)

    print("NER Output:\n", ner_tree)
```

```python

# FASTTEXT vs WORD2VEC COMPARISON


import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
import string
from gensim.models import Word2Vec, FastText

#  Download data
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('punkt_tab')


#  Prepare corpus

sentences = []

for fileid in gutenberg.fileids():
    text = gutenberg.raw(fileid)
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))

    for sent in text.split('\n'):
        words = word_tokenize(sent)
        if len(words) > 2:
            sentences.append(words)

print("Total sentences:", len(sentences))


# Train Word2Vec

w2v_model = Word2Vec(sentences, vector_size=100, window=5, min_count=2)


#  Train FastText

ft_model = FastText(sentences, vector_size=100, window=5, min_count=2)


#  Compare Similar Words

word = "king"

print("\nWord2Vec Similar Words:")
print(w2v_model.wv.most_similar(word))

print("\nFastText Similar Words:")
print(ft_model.wv.most_similar(word))


# Test Rare / Misspelled Word

test_word = "kings"  # or typo like "kingg"

print("\nWord2Vec (may fail):")
try:
    print(w2v_model.wv.most_similar(test_word))
except:
    print("Word not in vocabulary")

print("\nFastText (handles it):")
print(ft_model.wv.most_similar(test_word))
```
