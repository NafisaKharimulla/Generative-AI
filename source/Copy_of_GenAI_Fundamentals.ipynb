## 🌱 What is Generative AI?

Generative AI refers to models that can **generate new content**:
- Text
- Images
- Code
- Audio

Unlike traditional ML (predict/classify), GenAI **creates**.

### Examples:
- ChatGPT → text generation
- DALL·E → image generation
- GitHub Copilot → code completion

## 🧠 Large Language Models (LLMs)

LLMs are deep learning models trained on massive text corpora.

Examples:
- BERT → Understanding-focused
- GPT → Generation-focused
- T5 → Text-to-text framework

Key properties:
- Billions of parameters
- Trained using self-supervised learning
- Use Transformer architecture

## 🔄 Transformer Architecture

Transformers replaced RNNs/LSTMs by enabling:
- Parallel processing
- Long-range dependency capture

### Core Components:
- Token Embeddings
- Positional Encoding
- Multi-Head Self-Attention
- Feed Forward Networks

## 👀 Attention Mechanism (Self-Attention)

Self-attention allows each word to **attend to all other words**.

It computes:
- Query (Q)
- Key (K)
- Value (V)

This helps the model understand **context** and **relationships**.

## 🧠 Prompt Engineering

## 🎯 Prompting Techniques

1. **Zero-shot Prompting**
   - No examples

2. **One-shot Prompting**
   - One example

3. **Few-shot Prompting**
   - Multiple examples

## ⚠️ Model Limitations – Hallucinations

LLMs may:
- Generate incorrect facts
- Sound confident but be wrong

Mitigation:
- Grounding with data
- Retrieval-Augmented Generation (RAG)
- Explicit constraints in prompts

```python
from transformers import pipeline

# Load GPT-based model
generator = pipeline("text-generation", model="gpt2")

# Sample news article (you can replace this with real news)
article = """
Artificial Intelligence is rapidly transforming industries across the globe.
From healthcare to finance, AI-powered tools are improving efficiency, reducing costs,
and enabling better decision-making. Experts believe that AI adoption will continue
to grow, creating new opportunities while also raising concerns about job displacement
and ethical challenges.
"""

# Prompt for summarization
prompt = f"""
Summarize the following news article in 3-4 lines:

{article}

Summary:
"""

# Generate summary
output = generator(prompt, max_length=120, num_return_sequences=1)

print(output[0]['generated_text'])

```

```python
# Zero-shot vs Few-shot Prompting (Single Cell)

from transformers import pipeline

# Load text generation model
generator = pipeline("text-generation", model="gpt2")


#  Input Text

text = "AI is transforming industries by automating tasks and improving efficiency."


# Zero-shot Prompt

zero_shot_prompt = f"""
Summarize the following text into bullet points:

{text}
"""

zero_shot_output = generator(
    zero_shot_prompt,
    max_length=80,
    num_return_sequences=1
)


# Few-shot Prompt (with examples)

few_shot_prompt = f"""
Summarize the text into bullet points.

Example:
Text: Cloud computing stores data online.
Summary:
- Stores data online
- Accessible anywhere

Example:
Text: Machine learning finds patterns in data.
Summary:
- Finds patterns
- Improves predictions

Now:
Text: {text}
Summary:
"""

few_shot_output = generator(
    few_shot_prompt,
    max_length=120,
    num_return_sequences=1
)


# Print Results

print(" Zero-shot Output:\n")
print(zero_shot_output[0]['generated_text'])

print("\n" + "="*50 + "\n")

print(" Few-shot Output:\n")
print(few_shot_output[0]['generated_text'])
```

```python
# Identify Hallucinations
from transformers import pipeline

# Load model
generator = pipeline("text-generation", model="gpt2")

# Prompts to test hallucination
prompts = [
    "Who is the president of Mars?",
    "What is the capital of France?",
    "Who invented the Internet in 1800?"
]

# Generate outputs
for prompt in prompts:
    output = generator(prompt, max_length=50, num_return_sequences=1)[0]['generated_text']

    print(f"Prompt: {prompt}")
    print(f"Output: {output}")

    # Simple rule-based check
    if "Mars" in prompt or "1800" in prompt:
        print(" Possible Hallucination Detected")
    else:
        print(" Likely Correct")

    print("\n" + "-"*60 + "\n")
```
